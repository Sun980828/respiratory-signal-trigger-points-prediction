```python
from utils import listcsv
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
root_path = './Train_data'
csv_lists = listcsv(root_path)
```

```python
csv_lists[0], csv_lists[1] = csv_lists[1], csv_lists[0]
```

```python
DATA = []
for path in csv_lists:
    csv_data = pd.read_csv(path, header=None).to_numpy()
    DATA.append(csv_data)
    print(path)
    print(csv_data.shape)
print(len(DATA))
```

```python
num = 10

fig, ax1 = plt.subplots(len(DATA[num]), 1, figsize=(30, 5*len(DATA[num])))
k = 0
for signal in DATA[num]:
    time_steps = np.arange(len(signal))
    ax1[k].plot(time_steps, signal, marker = '.', markersize = 1, linestyle = '-')
    k += 1
```

```python
train_data = []
# test_data = []
for data in DATA:
    # i = np.random.randint(len(data))
    # test_data.append(data[i])
    # d = np.delete(data, i, axis=0)
    # print(d.shape)
    for x in data:
        train_data.append(x)
```
